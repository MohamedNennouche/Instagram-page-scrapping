```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
posts_df = pd.read_csv('vic_post_data.csv', index_col='upload_date', parse_dates=True)
posts_df.head()
```

On procÃ¨de Ã  la suppression d'un certain nombre de colonnes dont on n'a pas besoin

```python
posts_df.drop(['height', 'width', 'gating_info', 'fact_check_overall_rating', 'fact_check_information', 'sensitivity_friction_info', 'media_overlay_info', 'media_preview', 'username', 'full_name', 'profile_pic_url', 'profile_pic_url_hd', 'viewer_can_reshare', 'accessibility_caption', 'commenting_disabled_for_viewer', 'id', 'tracking_token', 'tagged_users', 'caption_is_edited', 'has_ranked_comments', 'comments_disabled', 'location'], axis='columns', inplace = True)
posts_df.head()
```

```python
posts_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
posts_df.columns
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure()
posts_df['likes'].plot(style='k.', c ='b', figsize=(16,10))
plt.ylim([0,800])
```

On va maintenant regarder la tendance en moyennant les mensuellement des j'aimes en moyennant la valeur du nombre de mention j'aime 

```python
plt.figure(figsize=(16,10))
posts_df['likes'].plot(style='k.', label="Nombre de mention j'aimes", c ='b')
posts_df['likes'].resample('M').mean().plot(label='Moyenne mensuelle', lw=3, ls=':', alpha = 0.8, c='r')
#posts_df['likes'].resample('W').mean().plot(label='Moyenne hebdomadaire', lw = 2, ls = '--' , alpha = 0.8)
plt.legend()
plt.ylim([0,800])
plt.show()
```

On peut Maintenant faire quelques graphiques plus reprÃ©sentatif vis-Ã -vis du nombre de mentions j'aime sur les posts

```python
moyenne_annee = posts_df['likes'].resample('Y').mean()
moyenne_mois = posts_df['likes'].resample('M').mean()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(moyenne_annee)
plt.xlabel("Les annÃ©es")
plt.ylabel("Nombre de mentions j'aimes en moyenne")
plt.subplot(1,2,2)
plt.plot(moyenne_mois)
plt.xlabel("Les mois")
plt.ylabel("Nombre de mentions j'aimes en moyenne")
plt.show()
```

Maintenant en faisant un comparatif sur les mois pour les 3 annÃ©es d'activitÃ©s complÃ¨tes du VIC on observe cela 

```python
moyenne_2021 = list(posts_df['2021']['likes'].resample('M').mean())
moyenne_2020 = list(posts_df['2020']['likes'].resample('M').mean())
moyenne_2019 = list(posts_df['2019']['likes'].resample('M').mean())
mois = range(1,13)
plt.figure(figsize=(12,8))
plt.plot(mois,moyenne_2021, label="Moyenne 2021")
plt.plot(mois,moyenne_2020, label="Moyenne 2020")
plt.plot(mois,moyenne_2019, label="Moyenne 2019")
plt.legend()
plt.xlabel("Les mois")
plt.ylabel("Nombre de mentions j'aimes en moyenne")
plt.show()
```
