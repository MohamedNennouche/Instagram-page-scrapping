# Viewing data from the Vision and Innovation Club's Instagram page
## Importing modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
posts_df = pd.read_csv('vic_post_data.csv', index_col='upload_date', parse_dates=True)
posts_df.head()
```

A number of columns that are not needed are deleted

```python
posts_df.drop(['height', 'width', 'gating_info', 'fact_check_overall_rating', 'fact_check_information', 'sensitivity_friction_info', 'media_overlay_info', 'media_preview', 'username', 'full_name', 'profile_pic_url', 'profile_pic_url_hd', 'viewer_can_reshare', 'accessibility_caption', 'commenting_disabled_for_viewer', 'id', 'tracking_token', 'tagged_users', 'caption_is_edited', 'has_ranked_comments', 'comments_disabled', 'location', 'Unnamed: 0'], axis='columns', inplace = True)
posts_df.head()
```

```python
posts_df.describe()
```

```python
posts_df.columns
```

```python
sns.set_theme()
plt.figure()
posts_df['likes'].plot(style='k.', c ='b', figsize=(16,10), xlabel='Upload date')
plt.ylim([0,800])
plt.title('Instagram page posts and number of likes')
```

On va maintenant regarder la tendance en moyennant les mensuellement des j'aimes en moyennant la valeur du nombre de mention j'aime 

```python
plt.figure(figsize=(16,10))
posts_df['likes'].plot(style='k.', label="Number of likes", c ='b', alpha=0.6)
posts_df['likes'].resample('3M').mean().plot(label='Quarterly average', lw=3, ls=':', alpha = 0.8, c='r')
posts_df['likes'].resample('Y').mean().plot(label='Annual average', lw=3, ls='--', c='g')
plt.legend()
plt.ylim([0,800])
plt.xlim(['2018','2022'])
plt.xlabel('Upload date')
plt.title("Average trends in the number of likes on the Instagram page")
```

Thus we see a clear evolution of the average number of mentions j'aime of the publications of the page instagram of the VIC, essentially during the year 2021, comparing the average at the beginning of 2021 and at its end it is more than double of annual average. 

We can now make some more representative graphs regarding the number of likes on the posts

```python
moyenne_annee = posts_df['likes'].resample('Y').mean()
moyenne_mois = posts_df['likes'].resample('M').mean()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(moyenne_annee)
plt.xlabel("Years")
plt.title("Annual average")
plt.subplot(1,2,2)
plt.plot(moyenne_mois)
plt.xlabel("Months")
plt.title("Mensual average")
plt.show()
```

Now if we make a comparison on the months for the 3 years of complete activities of the VIC we observe that 

```python
moyenne_2021 = list(posts_df['2021']['likes'].resample('M').mean())
moyenne_2020 = list(posts_df['2020']['likes'].resample('M').mean())
moyenne_2019 = list(posts_df['2019']['likes'].resample('M').mean())

moyenne_2021_trim = list(posts_df['2021']['likes'].resample('3M').mean())
moyenne_2020_trim = list(posts_df['2020']['likes'].resample('3M').mean())
moyenne_2019_trim = list(posts_df['2019']['likes'].resample('3M').mean())
mois = range(1,13)
trim = range(1, len(moyenne_2021_trim)+1)
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.plot(mois,moyenne_2021, label="2021 average")
plt.plot(mois,moyenne_2020, label="2020 average")
plt.plot(mois,moyenne_2019, label="2019 average")
plt.legend()
plt.xlabel("Months")
plt.title("Mensual average number of likes")
plt.subplot(1,2,2)
plt.plot(trim,moyenne_2021_trim, label="2021 average")
plt.plot(trim,moyenne_2020_trim, label="2020 average")
plt.plot(trim,moyenne_2019_trim, label="2019 average")
plt.legend()
plt.xlabel("Quarters")
plt.title("Quarterly average number of likes")
plt.show()
```

We confirm again with this comparison that the evolution of the page has been constant over the years, and clearly the year 2021 has been an excellent year in terms of communication through Instagram.

```python
plt.figure(figsize=(20,20))
plt.subplot(5,1,1)
plt.plot(reels["Upload_date"],reels["Reach"], label="Reach")
plt.plot(reels["Upload_date"],reels["Plays"], label="Plays")
plt.xlabel("Reels")
plt.ylabel("Number of accounts reached and views of reels")
plt.legend()
plt.subplot(5,1,2)
plt.plot(reels["Upload_date"],reels["Likes"])
plt.xlabel("Rreels")
plt.ylabel("Likes")
plt.subplot(5,1,3)
plt.plot(reels["Upload_date"],reels["Comments"])
plt.xlabel("Reels")
plt.ylabel("Comments")
plt.subplot(5,1,4)
plt.plot(reels["Upload_date"],reels["Shares"])
plt.xlabel("Reels")
plt.ylabel("Shares")
plt.subplot(5,1,5)
plt.plot(reels["Upload_date"],reels["Saves"])
plt.xlabel("Reels")
plt.ylabel("Saves")
plt.show()
```
